<H1>Visualisasi Data</H1>

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
data = pd.read_csv('data.csv')
data
```

```python
x = data['Ukuran']
y = data['Harga Sewa']

plt.title('Hubungan antara Ukuran dan Harga Sewa')
plt.plot(x, y)
plt.xlabel('Ukuran')
plt.ylabel('Harga Sewa')
plt.show()
```

```python
x = data['Tipe Bangunan']
y = data['Tarif Internet']

plt.title('Hubungan antara Tarif Internet dan Tipe Bangunan')
plt.scatter(x, y)
plt.xlabel('Tipe Bangunan')
plt.ylabel('Tarif Internet')
plt.show()
```

```python
tipe_a = data[data['Tipe Bangunan'] == 'A']
tipe_b = data[data['Tipe Bangunan'] == 'B']
tipe_c = data[data['Tipe Bangunan'] == 'C']

rate_sewa_a = sum(tipe_a['Harga Sewa'])/len(tipe_a['Harga Sewa'])
rate_sewa_b = sum(tipe_b['Harga Sewa'])/len(tipe_b['Harga Sewa'])
rate_sewa_c = sum(tipe_c['Harga Sewa'])/len(tipe_c['Harga Sewa'])

tipe = ['A','B','C']
rate = [rate_sewa_a,rate_sewa_b,rate_sewa_c]

plt.title('Tren Harga Sewa di setiap Tipe Bangunan')
plt.bar(tipe, rate)
plt.xlabel('Tipe Bangunan')
plt.ylabel('Harga Sewa')
plt.show()
```
